```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

data = pd.read_csv('diabetes.csv');
print('Pima Indian Diabetes Dataset loaded...')
i,a = data.shape

print('\nTotal instances available :',i)
print('\nTotal attributes present :',a-1)

print("\nFirst Five instances of dataset:") 
print(data.head(5))

y = data.iloc[:,-1].values
x = data.iloc[:,0:-1].values

i,a = x.shape
print('\nTotal instances available in x :',i)
print('\nTotal attributes present in x :',a-1)

print('\nTotal instances available in y:',i)
print('\nTotal attributes present in y:',1)

print("\nSplitting the data into train test split")
x_train,x_test,y_train,y_test = train_test_split(x,y)
print("\nTotal values of X train is:",x_train.shape)
print("\nTotal values of X test is:",x_test.shape)
print("\nTotal values of Y train is:",y_train.shape)
print("\nTotal values of Y test is:",y_test.shape)

clf = GaussianNB()
clf.fit(x_train,y_train)

print("\nAccuracy for the given dataset",clf.score(x_test,y_test))
```

```python
import numpy as np
import pandas as pd

from sklearn.naive_bayes import GaussianNB

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('sd.csv')

print(df)

x = df.iloc[:,0:-1]
y = df.iloc[:,-1]

le = LabelEncoder()

x1 = le.fit_transform(df.iloc[:,0])
x2 = le.fit_transform(df.iloc[:,1])
x3 = le.fit_transform(df.iloc[:,2])
y = le.fit_transform(df.iloc[:,3])

x = list(zip(x1,x2,x3))
#print(x)
x_train,x_test,y_train,y_test = train_test_split(x,y)
print(x_train)
clf = GaussianNB().fit(x_train,y_train)

print(clf.score(x_test,y_test))
newx = pd.read_csv('sdtest.csv')
print(newx)
newx1 = le.fit_transform(newx.iloc[:,0])
newx2 = le.fit_transform(newx.iloc[:,1])
newx3 = le.fit_transform(newx.iloc[:,2])
newx = list(zip(newx1,newx2,newx3))
print(newx)
ypred = clf.predict(newx)
print(ypred)
```
